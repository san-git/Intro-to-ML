In Linear Regression, we try to model the following,

$\begin{align*}
y = f(x_i)
\end{align*}$

Assuming f() is a linear expression.


Now, if g() is a non-linear function, such that,

$\begin{align*}
y = g(f(x_i))
\end{align*}$

Then we can manipulate the above as,

$\begin{align*}
g^{-1}(y) = f(x_i)
\end{align*}$

Hence, reducing a non-linear relationship to a linear one.


Since, we are trying to model a step function that varies between 0 and 1, we use the __sigmoid__ function, which is defined as follows-

$\begin{align*}
y = \frac{1}{1+e^{f(x)}}
\end{align*}$

Following the method above, 

$\begin{align*}
1+e^{f(x)} = \frac{1}{y}
\end{align*}$

$\begin{align*}
\frac{1-y}{y} = e^{f(x)}
\end{align*}$

$\begin{align*}
\ln{\left(\frac{1-y}{y}\right)} = f(x)
\end{align*}$

Since we are interested in modeling the probabilities of y being 1 or 0, we change y to probability of y=1,

$\begin{align*}
\ln{\left(\frac{1-P(y=1)}{P(y=1)}\right)} = f(x)
\end{align*}$

### Confusion Matrix
| Test\Actual | False | n | True | n | Total
|------|--------------|-------------|------------------------------------
|Negative | True Negative (TN) | d | False Negative (FN) | b | b + d
|Positive | False Positive (FP) | c | True Postive (TP) | a | a + c
|Total | | c + d | | a + b | a + b + c + d  

### Metrics in a confusion matrix

$\begin{align*}
Precision = \frac{a}{a + c}
\end{align*}$

Sensitivity or Recall 
$\begin{align*}
= \frac{a}{a + b}
\end{align*}$

$\begin{align*}
Specificity = \frac{d}{c + d}
\end{align*}$
